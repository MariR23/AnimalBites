### Outline
This notebook was created using deepnote and provides a data analysis of the Animal Bites dataset using BigQuery.
It also includes visualizations using Tableau.

The dataset contains animal bite incidents reported to and investigated by Louisville Metro Department of Public Health and Wellness  from 1985 to 2017. You can find this dataset on kaggle [here](https://www.kaggle.com/datasets/rtatman/animal-bites)

The business questions included in this notebook are:
* Which animal species are more likely to bite humans
* Does the gender of the animal affect the likelyhood of it biting and is it the case for every species


## Which animal species are more likely to bite humans
To answer this question, I created a table that counts the amount of animal bite incidents for each species.

```


df_1 = _deepnote_execute_sql("""SELECT DISTINCT 
  SpeciesIDDesc,
  COUNT(SpeciesIDDesc) AS Count_bites 
  
FROM 
  '/work/AnimalBites_v1.csv'

GROUP BY
  SpeciesIDDesc 
ORDER BY 
  Count_bites DESC
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_1
```

We can see that the species with the most reported bites are dogs, cats and bats.
Racoon, horses, ferrets, rabbits and skunk have a considerebly lower amount of bite instances.

## Does the gender of the animal affect the likehood of it biting

To determine the distribution of bites between genders, I created a table that sorts the bites by species and gender.

```


df_2 = _deepnote_execute_sql("""SELECT DISTINCT 
  SpeciesIDDesc,
  GenderIDDesc,
  COUNT(SpeciesIDDesc) AS Count_bites  --This creates a new column named Count_bites
  
FROM 
  '/work/AnimalBites_v1.csv'

WHERE
  GenderIDDesc IS NOT null AND
  SpeciesIDDesc IS NOT null AND
  GenderIDDesc NOT LIKE '%UNKNOWN%'

GROUP BY
  SpeciesIDDesc,
  GenderIDDesc 

ORDER BY 
  Count_bites DESC  --This orders the table according to the amount of bites per species in a descending order
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_2
```

We can conclude from this table that gender does affect the likehood of bites.
Male dogs and cats are more likely to bite humans than their female counterparts.
There is not enough data to make a conclusion when it comes to horses, skunks, ferrets, rabbits, racoons and bats.
